# How many sinle-path commodities using (4,10)


```python
import pandas as pd

excluded_commodities = [3, 5, 30, 39, 41, 42, 64, 66, 67, 102, 116, 117, 118, 138, 166, 167, 168, 169, 390, 393, 395, 398, 412, 415, 437, 440, 462, 465, 473]

total_count = 0

for o in range(1, 26):
    for d in range(1, 26):
      if o != d
        commodity_number = o + (d - 1) * 25
        if commodity_number not in excluded_commodities:
            sheet_name = f"{commodity_number}1"

            try:
                sigma_df = pd.read_excel("Sigma.xlsx", sheet_name, engine='openpyxl')

                if sigma_df.loc[10, 4] == 1 or sigma_df.loc[4, 10] == 1:
                    total_count += 1

            except FileNotFoundError:
                print(f"Sigma file for commodity {commodity_number} not found.")

print(f"Total number of commodities with llink(10,4) or (4,10) equal to 1: {total_count}")
```
