# Railroad-State


```python
import pandas as pd

data = pd.read_csv('TrainAccidents.csv')

unique_railroads = data['RAILROAD'].unique()
unique_states = data['STATE '].unique()

railroad_state_data = {'State': unique_states}

for railroad in unique_railroads:
    railroad_data = data[data['RAILROAD'] == railroad]

    railroad_states = []

    for state in unique_states:
        if state in railroad_data['STATE '].values:
            railroad_states.append(1)
        else:
            railroad_states.append(0)

    railroad_state_data[railroad] = railroad_states

railroad_state_table = pd.DataFrame(railroad_state_data)

railroad_state_table.set_index('State', inplace=True)

railroad_state_table.to_csv('railroad_state_table.csv')
```
