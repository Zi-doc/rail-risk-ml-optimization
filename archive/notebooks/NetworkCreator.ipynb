```python
import folium

m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

def calculate_distance(coord1, coord2):
    lat1, lon1 = map(float, coord1.split(', '))
    lat2, lon2 = map(float, coord2.split(', '))
    return ((lat1 - lat2) ** 2 + (lon1 - lon2) ** 2) ** 0.5

for railroad_name, states_list in connected_states.items():
    states_positions = [state_positions[state] for state in states_list if state in state_positions]

    if len(states_positions) == 1:
        state = states_list[0]
        print(f"{railroad_name} passes through {state}")
    elif len(states_positions) > 1:
        line_colors = ['blue', 'red', 'green', 'purple', 'orange', 'pink', 'brown', 'gray', 'black']
        color_index = 0

        path_list = [states_list[0]]
        remaining_states = states_list[1:]

        while remaining_states:
            last_state = path_list[-1]
            distances = [calculate_distance(last_state, state_positions[state]) for state in remaining_states]

            closest_index = distances.index(min(distances))
            closest_state = remaining_states[closest_index]

            path_list.append(closest_state)

            del remaining_states[closest_index]

            path_string = f"{railroad_name} starts from {states_list[0]}"

            for state in path_list[1:]:
                path_string += f", then it goes to {state}"

            print(path_string)

            locations = [
                [float(coord) for coord in path_list[-2].split(', ')],
                [float(coord) for coord in path_list[-1].split(', ')]
            ]
            folium.PolyLine(
                locations=locations,
                color=line_colors[color_index % len(line_colors)]
            ).add_to(m)
            color_index += 1

m.save('usa_map.html')
```
