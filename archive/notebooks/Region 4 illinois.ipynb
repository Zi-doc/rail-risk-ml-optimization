```python
import geopandas as gpd
import requests
from io import BytesIO
import zipfile

url = "https://www2.census.gov/geo/tiger/GENZ2021/shp/cb_2021_us_county_20m.zip"

response = requests.get(url)
with zipfile.ZipFile(BytesIO(response.content), 'r') as z:
    z.extractall("region_4_shapefile")  # Extract the contents to a directory

il_counties = gpd.read_file('region_4_shapefile/cb_2021_us_county_20m.shp')

region_4_counties = il_counties[il_counties['COUNTYFP'].isin(['031', '037', '043', '063', '089', '091', '093', '097', '111','197'] )]

```

```python
import geopandas as gpd
import folium

il_counties = gpd.read_file('region_4_shapefile/cb_2021_us_county_20m.shp')

il_counties = il_counties[il_counties['STATE_NAME'] == 'Illinois']

m = folium.Map(location=[41.8781, -87.6298], zoom_start=8)  # Adjust the center and zoom level as needed

folium.GeoJson(il_counties).add_to(m)

m.save('region_4_counties_map.html')
```

```python
import geopandas as gpd
import folium

il_counties = gpd.read_file('region_4_shapefile/cb_2021_us_county_20m.shp')

specified_counties = ['031', '037', '043', '063', '089', '091', '093', '097', '111', '197']  # Add '111' for McHenry County
region_4_counties = il_counties[
    (il_counties['STATEFP'] == '17') &  # Filter for Illinois (STATEFP 17)
    (il_counties['COUNTYFP'].isin(specified_counties))  # Filter for specified counties
]

m = folium.Map(location=[40.0, -89.0], zoom_start=7)  # Adjust the center and zoom level as needed

for idx, row in region_4_counties.iterrows():
    folium.GeoJson(row.geometry, style_function=lambda x: {'color': 'blue', 'weight': 2, 'fillOpacity': 0}).add_to(m)

m.save('illinois_region_4_counties_map.html')
```

```python
import geopandas as gpd
import folium

il_counties = gpd.read_file('region_4_shapefile/cb_2021_us_county_20m.shp')

specified_counties = ['031', '037', '043', '063', '089', '091', '093', '097', '111', '197']  # Add '111' for McHenry County
region_4_counties = il_counties[
    (il_counties['STATEFP'] == '17') &  # Filter for Illinois (STATEFP 17)
    (il_counties['COUNTYFP'].isin(specified_counties))  # Filter for specified counties
]

m = folium.Map(location=[41.8781, -87.6298], zoom_start=8, tiles='CartoDB positron')

for idx, row in region_4_counties.iterrows():
    folium.GeoJson(row.geometry, style_function=lambda x: {'color': 'blue', 'weight': 2, 'fillOpacity': 0}).add_to(m)

cities = {
    'Chicago': (41.8781, -87.6298),
    'Aurora': (41.7606, -88.3201),
    'Joliet': (41.5250, -88.0817),
    'Naperville': (41.7508, -88.1535),
    'Elgin': (42.0354, -88.2826),
    'Waukegan': (42.3636, -87.8448),
    'Cicero': (41.8456, -87.7539)
}

for city, coordinates in cities.items():
    folium.CircleMarker(
        location=coordinates,
        radius=5,
        color='red',
        fill=True,
        fill_color='red'
    ).add_to(m)

m.save('region_4_county_cities_map.html')
```
