```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

excel_file_path = 'Coordination.xlsx'  # Replace with the path to your Excel file
df = pd.read_excel(excel_file_path)

geolocator = Nominatim(user_agent="city_geocoder", timeout=10)

def get_city_name(row, max_retries=3):
    retries = 0
    while retries < max_retries:
        try:
            location = geolocator.reverse((row['Ycoordinate'], row['Xcoordinate']), language='en', addressdetails=True)

            if 'city' in location.raw['address']:
                return location.raw['address']['city']
            elif 'town' in location.raw['address']:
                return location.raw['address']['town']
            elif 'village' in location.raw['address']:
                return location.raw['address']['village']
            else:
                return location.raw['address'].get('suburb', None)
        except GeocoderTimedOut:
            retries += 1
    return None

df['City'] = df.apply(get_city_name, axis=1)

df.to_excel(excel_file_path, index=False)

print("City names added to the Excel file.")
```
