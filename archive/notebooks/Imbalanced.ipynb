```python
import pandas as pd
import openpyxl

sheet = wb['Train Accidents (2000 - 2020)']

data = pd.DataFrame(sheet.iter_rows(values_only=True, min_row=1, max_row=70458, min_col=2, max_col=45))

data.columns = data.iloc[0]

data = data[1:]

target_columns = ['CASINJ', 'CASKLD', 'ACCDMG', 'EVACUATE', 'CARSHZD', 'CARSDMG']

with pd.ExcelWriter('results.xlsx') as writer:
    for column in target_columns:
        value_counts = data[column].value_counts()

        percentages = (value_counts / len(data)) * 100

        column_result = pd.DataFrame({'Column': [column] * len(value_counts),
                                      'Value': value_counts.index,
                                      'Count': value_counts.values,
                                      'Percentage': percentages.values})

        column_result.to_excel(writer, sheet_name=column, index=False)

print("Results saved to results.xlsx")
```
