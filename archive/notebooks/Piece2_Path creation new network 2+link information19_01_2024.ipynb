```python
import networkx as nx
from copy import deepcopy
import pandas as pd

adj_matrix = deepcopy(adjacency_matrix)

city_names = ['92', '93', '94', '95', '96', '97', '98', '99', '101', '102', '103', '104',
              '105', '106', '107', '108', '109', '110', '111', '114', '115', '116',
              '117', '118', '119', '120', '121', '122', '123', '124', '126', '127',
              '128', '129', '130', '131', '132', '133', '134', '135', '136', '137',
              '138', '139', '140', '141', '142', '143', '144', '145', '146', '147',
              '148', '149', '150', '151', '152', '153', '154', '156', '158', '160',
              '161', '162', '163', '164', '165', '166', '167']

city_indices = {city: i for i, city in enumerate(city_names)}

G = nx.DiGraph()
for i in range(len(adj_matrix)):
    for j in range(len(adj_matrix[i])):
        if adj_matrix[i][j] == 1:
            G.add_edge(city_names[i], city_names[j])

def find_paths(graph, start, end, max_stops, current_path=[]):
    current_path = current_path + [start]

    if start == end and len(current_path) > 1:
        yield current_path
    if len(current_path) > max_stops:
        return

    for neighbor in graph.neighbors(start):
        if neighbor not in current_path:
            yield from find_paths(graph, neighbor, end, max_stops, current_path)

all_paths_info = {}

max_stops = 5

for start in city_names:
    all_paths_info[start] = {}
    for end in city_names:
        if start != end:
            all_paths_info[start][end] = {}
            paths = list(find_paths(G, start, end, max_stops))
            if paths:
                for i, path in enumerate(paths, 1):
                    commodity_indx = city_indices[start] * len(adj_matrix) + city_indices[end]

                    binary_matrix = [[0] * len(city_names) for _ in range(len(city_names))]
                    for k in range(len(path) - 1):
                        start_node_index = city_indices[path[k]]
                        end_node_index = city_indices[path[k + 1]]
                        binary_matrix[start_node_index][end_node_index] = 1

                    all_paths_info[start][end][i] = binary_matrix

origin = '150'
destination = '149'
path = 1

binary_matrix = all_paths_info[origin][destination][path]

df = pd.DataFrame(binary_matrix, columns=city_names, index=city_names)

excel_file_path = f"{origin}_{destination}_path{path}_binary_matrix.xlsx"
df.to_excel(excel_file_path)

print(f"Binary matrix saved to {excel_file_path}")
```
