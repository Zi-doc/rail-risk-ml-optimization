# FeatureSelectionGBoost


```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.ensemble import GradientBoostingRegressor  # Gradient Boosting for regression

data = pd.read_csv('datawithTime.csv')

Continuous_features_cols = ['CARS', 'TEMP', 'TRNSPD', 'TONS', 'POSITON1', 'HEADEND1', 'LOADF1', 'EMPTYF1', 'HIGHSPD', 'hour', 'minute']
categorical_features_cols = ['RAILROAD', 'YEAR', 'MONTH', 'DAY', 'STATE ', 'VISIBLTY', 'WEATHER',
                             'TYPEQ', 'TRKCLAS', 'TYPTRK', 'CAUSE', 'ACCTRK']

X = data[Continuous_features_cols + categorical_features_cols]
y = data['CASINJ']

model = GradientBoostingRegressor(n_estimators=100)  # You can adjust hyperparameters as needed

num_features_to_select = None  # Set the number of features you want to select
rfe = RFE(model, n_features_to_select=num_features_to_select)

rfe.fit(X, y)

feature_ranking = rfe.ranking_

selected_feature_names = [feature for feature, ranking in zip(X.columns, feature_ranking) if ranking == 1]

optimal_num_features = rfe.n_features_

final_model = GradientBoostingRegressor(n_estimators=100)  # Use the same hyperparameters as the original model
final_model.fit(X[selected_feature_names], y)

selected_feature_importances = final_model.feature_importances_

print("Selected Features:", selected_feature_names)
print("Optimal Number of Selected Features:", optimal_num_features)

for feature, importance in zip(selected_feature_names, selected_feature_importances):
    print(f"Feature: {feature}, Importance: {importance}")
```
