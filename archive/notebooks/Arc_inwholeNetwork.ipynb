```python

import pandas as pd
import numpy as np

matrix = pd.read_excel("ArcPercentage.xlsx", sheet_name="Sheet1", header=None, skiprows=1, usecols="B:Z", nrows=25)

matrix_array = matrix.values

flattened_matrix = matrix_array.flatten()

indices = np.argsort(flattened_matrix)[::-1]

sorted_values = flattened_matrix[indices]

rows, cols = np.unravel_index(indices, matrix_array.shape)
rows += 1
cols += 1

order_with_values = list(zip(rows, cols, sorted_values))

print(order_with_values)

matrix = pd.read_excel("ArcPercentage.xlsx", sheet_name="Sheet2", header=None, skiprows=1, usecols="B:Z", nrows=25)

matrix_array = matrix.values

flattened_matrix = matrix_array.flatten()

indices = np.argsort(flattened_matrix)[::-1]

sorted_values = flattened_matrix[indices]

rows, cols = np.unravel_index(indices, matrix_array.shape)
rows += 1
cols += 1

order_with_values = list(zip(rows, cols, sorted_values))

print(order_with_values)
```
