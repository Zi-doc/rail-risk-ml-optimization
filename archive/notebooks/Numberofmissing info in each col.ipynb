```python
import pandas as pd

data = pd.read_excel('FullData.xlsx')

data.to_csv('FullData.csv', index=False)

df = pd.read_csv('FullData.csv')
```

```python
data2 = df
```

```python

df.columns = [column.lower() for column in df.columns]

df = df.apply(lambda x: x.astype(str).str.lower())

df.replace(["na", "nan", "#nan"], "", inplace=True)

print(df.head())
```

```python
unique_values_count = df.nunique()

result = list(zip(unique_values_count.index.values, unique_values_count.values))

print("Column Name\tNumber of Unique Values")
for column_name, unique_count in result:
    print(f"{column_name}\t{unique_count}")
```

```python
df = data
df.replace(['NaN', '', '#NA'], pd.np.nan, inplace=True)

unique_values_count = df.nunique()

result = list(zip(unique_values_count.index.values, unique_values_count.values))

print("Number of Unique Values")
for column_name, unique_count in result:
    if not pd.isna(unique_count): # exclude NaN and blank values
        print(f"{unique_count}")
```
