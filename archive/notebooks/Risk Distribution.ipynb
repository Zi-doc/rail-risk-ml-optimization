```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
df = pd.read_excel('Analysis_RiskOnLinkFile.xlsx', sheet_name='Sheet1', header=None, usecols='B:Z', skiprows=72, nrows=25)
adj_matrix = df.values.tolist()

ddf = pd.read_excel('Analysis_RiskOnLinkFile.xlsx', sheet_name='Sheet1', header=None, usecols='C:AA', skiprows=30, nrows=25)
arc_loads = ddf.values.tolist()
matrix_load_max = [row[:] for row in arc_loads]

load_forward_df = pd.read_excel('Analysis_RiskOnLinkFile.xlsx', sheet_name='Sheet1', header=None, usecols='C:AA', skiprows=30, nrows=25)
load_backward_df = pd.read_excel('Analysis_RiskOnLinkFile.xlsx', sheet_name='Sheet1', header=None, usecols='C:AA', skiprows=30, nrows=25)
load_forward = load_forward_df.values.tolist()
load_backward = load_backward_df.values.tolist()

G = nx.Graph()

G.add_nodes_from(range(1, len(adj_matrix) + 1))  # Nodes labeled from 1 to n

for i in range(len(adj_matrix)):
    for j in range(i + 1, len(adj_matrix[i])):
        if adj_matrix[i][j] == 1 or adj_matrix[j][i] == 1:
            total_load = max(matrix_load_max[i][j],matrix_load_max[j][i])

            G.add_edge(i + 1, j + 1, load=total_load)  # Add 1 to node indices for labeling from 1

node_positions = {
    1: (2.5, 11.8),
    2: (2.8, 10.1),
    3: (1, 7.8),
    4: (2.9, 6.6),
    5: (0, 4),
    6: (5, 10.9),
    7: (5.5, 8.7),
    8: (3.7, 7.1),
    9: (4.2, 6),
    10: (2.9, 2.7),
    11: (4.8, 3),
    12: (6, 2.8),
    13: (3,1.5),
    14: (1.6,1),
    15: (7,0),
    16: (5.2,13.5),
    17: (5,12),
    18: (6.15,11.7),
    19: (7.9,11.2),
    20: (6.3,11),
    21: (9.5,9.8),
    22: (6.6,8.1),
    23: (8,6.5),
    24: (6.5,6),
    25: (6,5),
}

plt.figure(figsize=(8, 6))

nx.draw_networkx_nodes(G, node_positions, node_size=400, node_color='chartreuse', label=True)

node_labels = {node: node for node in G.nodes()}  # Create a mapping to label nodes from 1
nx.draw_networkx_labels(G, node_positions, labels=node_labels, font_size=10)

edge_loads = [data['load'] for _, _, data in G.edges(data=True)]

if edge_loads:  # Check if the list is not empty
    max_edge_load = max(edge_loads)

    edge_loads_normalized = [load / max_edge_load * 10 for load in edge_loads]  # Adjust multiplier to control thickness
    edge_loads_Unnormalized = [load/10000 for load in edge_loads]  # Adjust multiplier to control thickness
    print(edge_loads_normalized)
    print(edge_loads_Unnormalized)
else:
    edge_loads_normalized = []

thickest_edge_index = edge_loads_normalized.index(max(edge_loads_normalized))
thickest_edge = list(G.edges())[thickest_edge_index]

nx.draw_networkx_nodes(G, node_positions, node_size=400, node_color='chartreuse', label=True)

node_labels = {node: node for node in G.nodes()}  # Create a mapping to label nodes from 1
nx.draw_networkx_labels(G, node_positions, labels=node_labels, font_size=10)

nx.draw_networkx_edges(G, node_positions, width=edge_loads_normalized, edge_color='gray', alpha=0.7)

nx.draw_networkx_edges(G, node_positions, edgelist=[thickest_edge], width=max(edge_loads_normalized),
                       edge_color='red', alpha=0.7)

plt.savefig('0Risk.png')

plt.show()
```
