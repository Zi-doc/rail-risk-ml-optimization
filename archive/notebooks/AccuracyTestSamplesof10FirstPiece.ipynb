```python
import pandas as pd
import numpy as np
import networkx as nx

excel_file = pd.ExcelFile('Cost_matrix.xlsx')
cost_matrix = excel_file.parse('Sheet1', header=None, skiprows=2, usecols="C:AA", nrows=25).values

cost_matrix = np.array(cost_matrix)

adjacency_matrix = np.where(cost_matrix > 0, 1, 0)

selected_nodes = [1, 2, 3, 4, 5, 14, 8, 9, 10, 13]

subgraph = adjacency_matrix[np.ix_(selected_nodes, selected_nodes)]

digraph = nx.DiGraph(subgraph)

def find_all_paths(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return [path]
    paths = []
    for node in graph[start]:
        if node not in path:
            new_paths = find_all_paths(graph, node, end, path)
            for new_path in new_paths:
                paths.append(new_path)
    return paths

for source in selected_nodes:
    for target in selected_nodes:
        if source != target:
            source_node = source
            target_node = target
            print(f"Paths between node {source_node} and node {target_node}:")
            if target_node in digraph:
                paths = find_all_paths(digraph, source_node, target_node)
                if paths:
                    for path in paths:
                        selected_path = [selected_nodes[node-1] for node in path]
                        print(selected_path)
                else:
                    print("No path found.")
            else:
                print("Target node not reachable.")
            print()
```
